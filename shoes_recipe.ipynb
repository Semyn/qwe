```python
class Shoes:
    def __init__(self, type_but, gender, shoe_type, color, price, manufacturer, size):
        self.type_but = type_but       
        self.gender = gender
        self.shoe_type = shoe_type       
        self.color = color
        self.price = price
        self.manufacturer = manufacturer
        self.size = size

    def get_info(self):
        return {
            "Тип обуви": self.type_but,
            "Пол": self.gender,
            "Вид обуви": self.shoe_type,
            "Цвет": self.color,
            "Цена": self.price,
            "Производитель": self.manufacturer,
            "Размер": self.size
        }

    def update_info(self, **kwargs):
        for key, value in kwargs.items():
            if hasattr(self, key):
                setattr(self, key, value)

class ShoesController:
    def __init__(self, model):
        self.model = model

    def get_shoes_info(self):
        return self.model.get_info()

    def update_shoes_info(self, **kwargs):
        self.model.update_info(**kwargs)

class ShoesView:
    def display(self, info):
        print("Информация об обуви:")
        for key, value in info.items():
            print(f"{key}: {value}")

shoes = Shoes("мужская", "мужская", "кроссовки", "черный", 5000, "Nike", 42)

controller = ShoesController(shoes)

view = ShoesView()

info = controller.get_shoes_info()
view.display(info)
controller.update_shoes_info(color="синий", price=5500)

info = controller.get_shoes_info()
view.display(info)
```

```python
class Recipe:
    def __init__(self, name, author, type_recipe, description, video_link, ingredients, cuisine):
        self.name = name
        self.author = author
        self.type_recipe = type_recipe      
        self.description = description
        self.video_link = video_link
        self.ingredients = ingredients            
        self.cuisine = cuisine                    

    def get_info(self):
        return {
            "Название рецепта": self.name,
            "Автор": self.author,
            "Тип рецепта": self.type_recipe,
            "Описание": self.description,
            "Ссылка на видео": self.video_link,
            "Ингредиенты": self.ingredients,
            "Кухня": self.cuisine
        }

    def update_info(self, **kwargs):
        for key, value in kwargs.items():
            if hasattr(self, key):
                setattr(self, key, value)

class RecipeController:
    def __init__(self, model):
        self.model = model

    def get_recipe_info(self):
        return self.model.get_info()

    def update_recipe_info(self, **kwargs):
        self.model.update_info(**kwargs)

class RecipeView:
    def display(self, info):
        print("Информация о рецепте:")
        for key, value in info.items():
            if isinstance(value, list):
                print(f"{key}: {', '.join(value)}")
            else:
                print(f"{key}: {value}")

recipe = Recipe(
    name="Борщ",
    author="Иван Иванов",
    type_recipe="Первое блюдо",
    description="Традиционный украинский борщ с мясом и свеклой.",
    video_link="https://youtube.com/borch-recipe",
    ingredients=["свекла", "капуста", "картофель", "мясо", "морковь", "лук"],
    cuisine="Украинская"
)

recipe_controller = RecipeController(recipe)

recipe_view = RecipeView()

recipe_info = recipe_controller.get_recipe_info()
recipe_view.display(recipe_info)

recipe_controller.update_recipe_info(ingredients=["свекла", "капуста", "картофель", "мясо", "морковь", "лук", "чеснок"])

recipe_info = recipe_controller.get_recipe_info()
recipe_view.display(recipe_info)
```
