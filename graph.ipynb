```python
import networkx as nx
import matplotlib.pyplot as plt


G = nx.Graph()


G.add_node("Склад A")
G.add_node("Склад B")
G.add_node("Магазин 1")
G.add_node("Магазин 2")
G.add_node("Магазин 3")

G.add_edge("Склад A", "Магазин 1", weight=5)
G.add_edge("Склад A", "Магазин 2", weight=10)
G.add_edge("Склад A", "Склад B", weight=2)
G.add_edge("Склад B", "Магазин 2", weight=3)
G.add_edge("Склад B", "Магазин 3", weight=4)
G.add_edge("Магазин 1", "Магазин 3", weight=8)
G.add_edge("Магазин 2", "Магазин 3", weight=7)

start_node = "Склад A"
end_node = "Магазин 3"
path = nx.shortest_path(G, source=start_node, target=end_node, weight='weight')
path_cost = nx.path_weight(G, path, weight='weight')

print("Минимальный путь от", start_node, "до", end_node, ":", path)
print("Общая стоимость маршрута:", path_cost)

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=12, font_color='black')
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.title("Граф доставки товаров")
plt.show()

G.add_node("Склад C")
G.add_edge("Склад C", "Магазин 1", weight=6)
G.add_edge("Склад C", "Магазин 3", weight=2)

path_after_addition = nx.shortest_path(G, source=start_node, target=end_node, weight='weight')
path_cost_after_addition = nx.path_weight(G, path_after_addition, weight='weight')

print("После добавления нового склада:")
print("Минимальный путь от", start_node, "до", end_node, ":", path_after_addition)
print("Общая стоимость маршрута:", path_cost_after_addition)
```
